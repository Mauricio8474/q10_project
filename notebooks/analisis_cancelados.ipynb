# ETL

```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/cancelados.csv"
)

df.head(10)
```

```python
print(df.columns)
df.drop(columns=['Codigo_estudiante', 'Abreviatura_tipo_identificacion',
       'Numero_identificacion', 'Nombre_completo_estudiante',
       'Codigo_matricula', 'Codigo_programa',
       'Consecutivo_sede_jornada', 'Nombre_sede_jornada',
       'Consecutivo_periodo','Codigo_nivel',
       'Consecutivo_tipo_cancelacion',
       'Descripcion_tipo_cancelacion', 'Consecutivo_causa_cancelacion',
       'Codigo_administrativo', 'Nombre_administrativo', 'Fecha_nacimiento'], 
       inplace=True)
df.head(5)
```

```python
def separar_sede_programa(nombre):
    partes = nombre.split(' - ')
    sede = partes[0].strip()
    
    # Si hay 3 partes, la del medio es jornada (ignorar)
    # Si hay 2 partes, directo sede - programa
    programa = partes[-1].strip()
    
    return sede, programa

df[['Sede', 'Nombre_programa_limpio']] = df['Nombre_programa'].apply(
    lambda x: pd.Series(separar_sede_programa(x))
)
df.drop(columns=['Nombre_programa'], inplace=True)
df.head()
```

```python
# Convertir a datetime, manejando errores
df['Fecha_cancelacion'] = pd.to_datetime(df['Fecha_cancelacion'], errors='coerce')
df.head()
```

```python

print(df['Nombre_causa_cancelacion'].unique())
df['Observacion_cancelacion'].unique()
```

```python
import re

def clasificar_causa(observacion):
    if pd.isna(observacion) or str(observacion).strip() == '':
        return 'Sin información'
    
    obs = str(observacion).lower()
    
    # Orden importa: de más específico a más general
    if any(p in obs for p in ['circular 011', 'resolución 011']):
        return 'Cancelación administrativa'
    
    if any(p in obs for p in ['oncolog', 'enfermedad', 'salud', 'saludo', 'calamidad']):
        return 'Salud'
    
    if any(p in obs for p in ['matern', 'hija', 'hijo', 'cabeza de hogar', 'familiar']):
        return 'Responsabilidad familiar'
    
    if any(p in obs for p in ['otra universidad', 'otra institución', 'admitida en otra', 'cambio de programa', 'cambio de carrera', 'otro programa']):
        return 'Cambio de institución o programa'
    
    if any(p in obs for p in ['cambio de ciudad', 'se va de la ciudad', 'ausentarme', 'irse de la ciudad', 'cambió de ciudad']):
        return 'Cambio de ciudad'
    
    if any(p in obs for p in ['laboral', 'trabajo', 'horario', 'jornada laboral', 'económic', 'financier']):
        return 'Razones laborales o económicas'
    
    if any(p in obs for p in ['aplaz', 'congela', 'pausa', 'postergar']):
        return 'Aplazamiento'
    
    if any(p in obs for p in ['falta de interés', 'falta de interes', 'no le gusta', 'no cumple', 'gratuidad']):
        return 'Deserción por desinterés o requisitos'
    
    if any(p in obs for p in ['personal', 'voluntari', 'desiste', 'retiro']):
        return 'Motivos personales'
    
    return 'Otro'


df['Causa_definitiva'] = df['Observacion_cancelacion'].apply(clasificar_causa)
```

### Bloque 8 — Evolución por fecha (3 cortes)

```python
import pandas as pd

# ── Grupos con fechas diferenciadas ──────────────────
PROGRAMAS_GRUPO_B = ['TECNOLOGÍA EN GESTIÓN DE PRODUCCIÓN DE MODAS']
SEDES_GRUPO_B     = ['MINCA', 'BURITACA']

# ── Rangos por grupo ──────────────────────────────────
CORTES_A = pd.to_datetime([
    '2026-02-01', '2026-04-01', '2026-05-13', '2026-06-25'
])
ETIQUETAS_A = [
    'Seguimiento 1',
    'Seguimiento 2',
    'Seguimiento 3',
]

CORTES_B = pd.to_datetime([
    '2026-04-06', '2026-05-10', '2026-06-13', '2026-07-26'
])
ETIQUETAS_B = [
    'Seguimiento 1',
    'Seguimiento 2',
    'Seguimiento 3',
]

# ── Identificar a qué grupo pertenece cada fila ───────
df['Fecha_cancelacion'] = pd.to_datetime(df['Fecha_cancelacion'], errors='coerce')

es_grupo_b = (
    df['Nombre_programa_limpio'].isin(PROGRAMAS_GRUPO_B) |
    df['Sede'].isin(SEDES_GRUPO_B)
)

# ── Asignar seguimiento según grupo ───────────────────
def asignar_seguimiento(fecha, cortes, etiquetas):
    if pd.isna(fecha):
        return None
    for i in range(len(etiquetas)):
        if cortes[i] <= fecha < cortes[i + 1]:
            return etiquetas[i]
    return None

df['Seguimiento'] = [
    asignar_seguimiento(
        row['Fecha_cancelacion'],
        CORTES_B if es_grupo_b.loc[idx] else CORTES_A,
        ETIQUETAS_B if es_grupo_b.loc[idx] else ETIQUETAS_A
    )
    for idx, row in df.iterrows()
]

# ── Verificación ──────────────────────────────────────
print("Grupo A (programas y sedes regulares):")
print(df[~es_grupo_b]['Seguimiento'].value_counts().reindex(ETIQUETAS_A))

print("\nGrupo B (Modas + Minca + Buritaca):")
print(df[es_grupo_b]['Seguimiento'].value_counts().reindex(ETIQUETAS_B))

print(f"\nSin seguimiento asignado: {df['Seguimiento'].isna().sum()}")
```
